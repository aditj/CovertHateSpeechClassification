```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/logs/cumm_var_reduce.csv")
```

```python
df.drop(['date','n_succ','n_rounds'],axis=1,inplace=True)
```

```python
df_grouped = df.dropna().groupby(['n_dev','n_class_eav','prop_data_eav','class_dist_eav','greedy']).agg(['mean','std'])
```

```python
df_grouped
```

```python
final_df = df_grouped.groupby(['n_dev','n_class_eav','prop_data_eav','class_dist_eav']).apply(lambda x: pd.Series({'greedy_eav_acc':x['eaves_acc'].values[1][0],
                                                                                                                'greedy_eav_acc_std':x['eaves_acc'].values[1][1],
                                                                                                                'optimal_eav_acc':x['smart_eaves_accuracy'].values[0][0],                                                                                 'optimal_eav_acc_std':x['smart_eaves_accuracy'].values[0][1],
                                                                                                                'greedy_learner_acc_mean':x['learner_acc'].values[0][0],
                                                                                                                'greedy_learner_acc_std':x['learner_acc'].values[0][1],
                                                                                                                'optimal_learner_acc_mean':x['learner_acc'].values[1][0],
                                                                                                                'optimal_learner_acc_std':x['learner_acc'].values[1][1],}))
```

```python
final_df = final_df.reset_index().sort_values(by=['n_dev','n_class_eav','prop_data_eav','class_dist_eav'],ascending=[True,True,True,False])
```

```python
final_df
```

```python
final_df.to_csv("../data/logs/cummulative_experiment_results_final.csv")
```
